```python
data_country=pd.read_excel('Country-Code.xlsx')
```

```python
data=pd.merge(data,data_country,on='Country Code',how='left')
```

```python
data.Country.value_counts()
```

```python
data.columns
```

```python
ratings=data.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

# observations
1. when rating is b/w 4.5 to 4.9---->Excellent
2. when rating is b/w 4.0 to 3.4---->very good
3. when rating is b/w 3.5 to 3.9---->good
4. when rating is b/w 3.0 to 3.4---->average
5. when rating is b/w 2.0 to 2.4---->poor
6. max no. of people has not given reviews

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

```python
#palette is for mapping colors
#observations 1.max no. are not rated 2. max no. of rated are in 2.5 to 3.4
```

```python
#count plot freqcount
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
